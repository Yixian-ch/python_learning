## A naive search

```python
from random import randint, choice, seed
seed(0)
k = 100000

l = [randint(0,k) for _ in range(10000)]

def search(el,lst):
    count = 0
    for i in range(len(lst)):
        count += 1
        if el == lst[i]:
            return True, count
    return False, count


for _ in range(10):
    r = randint(0, k)
    inside, count = search(r,l)
    print(r, r in l, inside, count)
```

## Dichotomy search

```python
def dichotomy(el, lst):
    count = 0
    low = 0
    high = len(lst) - 1

    count += 1
    # 每一次只能比一种大小
    if lst[low] > el:
        return False, count
    count  += 1

    if lst[high] < el:
        return False, count
    count += 1

    if lst[low] == el:
        return True, count
    count += 1

    if lst[high] == el:
        return True, count

    # as list contains more than one element, high is bigger than low
    while low < high - 1:

        count += 1
        mid = (low + high) // 2
        if el == lst[mid]:
            return True, count

        count +=1
        if el < lst[mid]:
            high = mid

        else:
            low = mid

    return False, count


l.sort() # we sort l, so that dichotomy can kick in

for _ in range(10):
    r = randint(0, k) # about 1/1000 chance that we pick an element in the list by random
    inside, comp = dichotomy(r, l)
    print(r, r in l, inside, comp, sep='\t')
```

```python
def dichotomy_with_position(ls, ele):
    left = 0
    right = len(ls) - 1
    if ls[0] == ele:
        return True, 0

    if ls[-1] == ele:
        return True, len(ls)-1


    while left < right -1:
        mid = (right + left) // 2

        if ls[mid] == ele:
            return True, mid

        if ele < ls[mid]:
            right = mid

        else:
            left = mid

print(dichotomy_with_position(l, 197), l.index(197))
```

```python

from random import randint, seed, choice
from math import log2

seqs = ['abcdefgh',
        '1234567890',
        "a1b2c3d4f5",
        [0,1,2,3,4,5,6],
        ['a', 'bc', 'def'],
        [0, 'a', '2', 4, 'cd5'],
        (0,1,2,3),
        ('ab', 2,'34')]

# 1.
print('1.')
for seq in seqs:
    print(seq, str(seq), repr(seq), sep='\t')
    # by default print uses str() if available but str() and repr() are different
    # repr() is used to represent values as they should appear in python code to make it work
    # see how it is important for strings more than for numbers

    print('\t'.join(str(x) for x in seq))
    print('\t'.join(repr(x) for x in seq))
    print()
```

```python
# 2. a bit of complexity
# let there be the following functions taking an integer (or float) as input
print('2.')

# lambda is used to defined functions in a simple line, useful when you don't want to name your function
# these function are all monotonous and only have one intersection

a = lambda x: 4 * x ** 2 + 25 * x
b = lambda x: 2 * x ** 2 + 100 * x
c = lambda x: 1 * x * log2(x+1) + 128 * x
d = lambda x: 0.1 * x ** 3 + 0.5* x
funs = [a, b, c, d]
# here we want to compare the growth rate of these functions by combining them together, total possibilily of combinaison is !(len(funcitons)-1)
pairs = []
for i in range(len(funs)):
    for j in range(i+1, len(funs)):
        pairs.append((funs[i], funs[j]))
print(len(pairs))
```

```python
def croiss_point(f,g):
    count = 0
    i = 1
    order = f(i) > g(1)
    count += 2
    if order:
        print(f"value={i} f(1) > g(1) {f(i)} {g(i)}")
    else:
        print(f"value={i} f(1) <= g(1) {f(i)} {g(i)}")
    count +=2
    while order == (f(i) > g(i)):
        i += 1
        count += 2

    return i,count

for f,g in pairs:
    i,count = croiss_point(f,g)
    print(f"value={i-1} {f(i-1)} {g(i-1)} difference: count={count} value={i} {f(i)} {g(i)}")
```
